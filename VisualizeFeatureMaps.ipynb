# Import Libraries

```python
import keras
import cv2
import os
import numpy as np
from keras.models import load_model, Model
from keras import backend as K
from keras.utils import np_utils
from matplotlib import pyplot as plt
import matplotlib
matplotlib.use('TkAgg')
```

# Visualize Feature Maps

```python
# Read image you want to display its feature maps
img = cv2.imread('E:\\HOCTAP\DATN\\Dataset\\Image130_resize\\image149 (2).jpg')
img = img.reshape(-1,66,200,3)
cwd = os.getcwd()

model_dir =os.path.join(cwd, "Model_fold_9_Epoch_0002.h5")
print(model_dir)
model = load_model(model_dir)

layer_outputs = [layer.output for layer in model.layers[:9]] 

activation_model = Model(inputs=model.input, outputs=layer_outputs)

activations = activation_model.predict(img) 

first_layer_activation = activations[0]
second_layer_activation = activations[1]
third_layer_activation = activations[2]
fourth_layer_activation = activations[3]
fifth_layer_activation = activations[4]

fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
plt.title("Original Image")
ax1.imshow(img[0])
ax2 = fig.add_subplot(3,2,2)
plt.title("The First Feature Map")
ax2.matshow(first_layer_activation[0, :, :, 4], cmap='viridis')
ax3 = fig.add_subplot(3,2,3)
plt.title("The Second Feature Map")
ax3.matshow(second_layer_activation[0, :, :, 4], cmap='viridis')
ax4 = fig.add_subplot(3,2,4)
plt.title("The Third Feature Map")
ax4.matshow(third_layer_activation[0, :, :, 4], cmap='viridis')
ax5 = fig.add_subplot(3,2,5)
ax5.matshow(fourth_layer_activation[0, :, :, 4], cmap='viridis')
plt.title("The Fourth Feature Map")
ax6 = fig.add_subplot(3,2,6)
ax6.matshow(fifth_layer_activation[0, :, :, 4], cmap='viridis')
plt.title("The fifth Feature Map")
plt.show()
```
